```python
import os
from github import Github
from google.colab import files
import requests
from requests.structures import CaseInsensitiveDict

# Function to get CSV file from user via Google Colab
def upload_csv():
    # Prompt the user to upload a file
    uploaded = files.upload()  # This will prompt the user to upload the file

    # Extract the file name and content from the uploaded file
    for file_name in uploaded.keys():
        print(f"Uploaded file: {file_name}")
        return file_name

# Function to upload file to GitHub
def upload_to_github(file_path, repo_name, github_token):
    # Initialize GitHub instance using your token
    g = Github(github_token)

    # Get the authenticated user's repositories
    user = g.get_user()

    # Check if the repository exists
    try:
        repo = user.get_repo(repo_name)
        print(f"Repository '{repo_name}' found.")
    except:
        # If not found, create the repository
        repo = user.create_repo(repo_name)
        print(f"Created repository: {repo_name}")

    # Get the file content
    with open(file_path, 'r') as file:
        content = file.read()

    # Get the filename from the file path
    filename = os.path.basename(file_path)

    # Check if the file already exists in the repo
    try:
        file = repo.get_contents(filename)
        repo.update_file(file.path, f"Updated {filename}", content, file.sha)
        print(f"Updated existing file: {filename}")
    except:
        repo.create_file(filename, f"Added {filename}", content)
        print(f"Uploaded new file: {filename}")

    # Construct the URL to access the raw file
    url = f"https://raw.githubusercontent.com/{user.login}/{repo.name}/main/{filename}"
    return url

if __name__ == "__main__":
    # Get your GitHub token and repository name
    GITHUB_TOKEN = "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"
    REPO_NAME = "ATMECS_dataset"

    # Upload a CSV file
    csv_file = upload_csv()

    if csv_file:
        # Upload to GitHub if a CSV file was selected
        dataset_url = upload_to_github(csv_file, REPO_NAME, GITHUB_TOKEN)
        print(f"Dataset URL: {dataset_url}")
```
